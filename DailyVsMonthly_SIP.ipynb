```python
import pandas as pd
import yfinance as yf
from yahoofinancials import YahooFinancials
import plotly.express as px
```

```python
data = yf.download("^NSEI")
```

```python
nse_data_daily = pd.DataFrame()
```

```python
data
```

```python
nse_data_daily['Date'] = data.index
nse_data_daily['Close'] = data['Close'].values
```

```python
nse_data_daily.head()
```

```python
nse_data_daily.isnull().sum()
```

* Perfect no null values

```python
fig = px.line(nse_data_daily, x="Date", y="Close", title='NSE whole historical Data Daily')
fig.update_yaxes(title_text='Points') #Update Y-Axis title
fig.update_traces(line_color='#FF0000') #Update Chart colo
fig.show()
```

### Find Average Daily Percentage Increase

```python
close_list = nse_data_daily['Close'].values

ages = pd.Series(close_list)
changes = ages.pct_change() * 100

nse_data_daily['Percentage_Increase'] = changes
nse_data_daily['Percentage_Increase'].fillna(0, inplace=True)
```

```python
nse_data_daily
```

```python
fig = px.line(nse_data_daily, x="Date", y="Percentage_Increase", title='NSE whole historical Data Daily Increase')
fig.update_yaxes(title_text='Increase By') #Update Y-Axis title
fig.update_traces(line_color='#FF0000') #Update Chart colo
fig.show()
```

```python
daily_increase_average = nse_data_daily['Percentage_Increase'].mean()
print("The Daily Increase Percentage Average in NSE: ", daily_increase_average)
```

### Calculate the return on the Daily SIP till Date i.e 15 years

```python
sip_amount_daily = 500

principle_amount_daily = len(nse_data_daily)*sip_amount_daily

print("Total Amount you would have invested: ", principle_amount_daily)
daily_return = []

return_on_investment_daily = 0
for value in nse_data_daily['Percentage_Increase']:
#     print(value)
    return_on_investment_daily += sip_amount_daily + (return_on_investment_daily * (value/100))
    
print("Total Money in Account: ",return_on_investment_daily)


print("Return on Investment: ", return_on_investment_daily - principle_amount_daily)
```

# For Monthly

```python
data = yf.download("^NSEI", interval="1mo")

nse_data_monthly = pd.DataFrame()
nse_data_monthly['Date'] = data.index
nse_data_monthly['Close'] = data['Close'].values

nse_data_monthly
```

```python
fig = px.line(nse_data_monthly, x="Date", y="Close", title='NSE whole historical Data Monthly')
fig.update_yaxes(title_text='Points') #Update Y-Axis title
fig.update_traces(line_color='#FF0000') #Update Chart colo
fig.show()
```

### Find Average Monthly Percentage Increase

```python
close_list = nse_data_monthly['Close'].values

ages = pd.Series(close_list)
changes = ages.pct_change() * 100

nse_data_monthly['Percentage_Increase'] = changes
nse_data_monthly['Percentage_Increase'].fillna(0, inplace=True)

fig = px.line(nse_data_monthly, x="Date", y="Percentage_Increase", title='NSE whole historical Data Monthly Increase')
fig.update_yaxes(title_text='Increase By') #Update Y-Axis title
fig.update_traces(line_color='#FF0000') #Update Chart colo
fig.show()
```

```python
monthly_increase_average = nse_data_monthly['Percentage_Increase'].mean()
print("The Monthly Increase Percentage Average in NSE: ", monthly_increase_average)
```

### Calculate the return on the Monthly SIP till Date i.e 15 years

```python
sip_amount_daily = 500
sip_amount_monthly = sip_amount_daily*20.324025 #matching the daily sip figure.


principle_amount = len(nse_data_monthly)*sip_amount_monthly

print("Total Amount you would have invested: ", principle_amount)

return_on_investment_monthly = 0
for value in nse_data_monthly['Percentage_Increase']:
#     print(value)
    return_on_investment_monthly += sip_amount_monthly + (return_on_investment_monthly * (value/100))
    
print("Total Money in Account: ",return_on_investment_monthly)


print("Return on Investment: ", return_on_investment_monthly - principle_amount)
```

### 

```python
print("Return on Investment Daily: ", return_on_investment_daily)
print("Return on Investment Monthly: ", return_on_investment_monthly)

print("Difference: ", return_on_investment_daily - return_on_investment_monthly)
```
